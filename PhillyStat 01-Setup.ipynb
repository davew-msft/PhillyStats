# Poverty Profiteering

Poverty profiteering, the practice of exploiting the economically disadvantaged for financial gain, is a critical issue plaguing the City of Philadelphia. Poverty profiteering manifests in various forms, including predatory lending, exploitative housing practices, and inadequate access to essential services. It brings blight and disorder.  

LLC and property owners buy and hold a property with no intention of being a "neighbor".  In some cases the LLC operating agreement even states this intent.  

This is an _intentional business model_.  Buying or investing in a property and holding it hoping to make money from it is just the way the world works and there's nothing wrong with that.  But when it becomes intentional to make any sort of investment in the property, keeping up with your liabilities, etc, is flouting the law.

# Carto 
Using carto we can query a lot of phila.gov.
This is the helper function:


```python
# Carto query setup...pass in a query and get back a sql table

baseURL = "https://phl.carto.com/api/v2/sql?q="

def ReadCarto (qry,tblName):
    response = requests.get(baseURL + qry)
    rawData = response.content.decode("utf-8")

    RawDF = (spark.read
        .option("multiline", True)
        .json(sc.parallelize([rawData])))

    # this gets the 4 nested datasets
        # fields:  fields list and datatypes
        # rows:  the actual data
        # time:  exec time
        # total_rows
    # display(RawDF)
    RawDF.select("total_rows").show()
    print("Carto query:" + baseURL + qry)
    #RawDF.printSchema()

    # now just explode out the rows struct, that's all we care about
    explodedDF = (
                RawDF
                    .withColumn("results",explode(RawDF.rows))
                    .select ("results.*"))
    explodedDF.createOrReplaceTempView(tblName)
    return explodedDF
```

```python
# this version handles loading large tables in bulk
# this is beyond kludgey.  

def ReadCartoTable (tbl,batchSize):
    if DownloadLatestData != 1:
        print("Not rebuilding table this run")
        return
    print ("starting:" + tbl)
    # 1 get the min/max cartodb_id
    finalURL = baseURL + "select min(cartodb_id) as min, max(cartodb_id) as max, count(*) as cnt from " + tbl
    response = requests.get(finalURL)
    rawData = response.content.decode("utf-8")

    RawDF = (spark.read
        .option("multiline", True)
        .json(sc.parallelize([rawData])))

    # now just explode out the rows struct, that's all we care about
    explodedDF = (
                RawDF
                    .withColumn("results",explode(RawDF.rows))
                    .select ("results.*"))
    #explodedDF.show()

    minID = int(explodedDF.select("min").collect()[0][0])
    print(minID)
    maxID = int(explodedDF.select("max").collect()[0][0])
    print(maxID)
    cnt = int(explodedDF.select("cnt").collect()[0][0])
    print(cnt)

    # 2 use that to make paginated calls
    allDataDF = None
    currentID = minID
    while currentID <= maxID:
        rangeEndID = currentID + batchSize
        finalURL = baseURL + "select * from " + tbl + " WHERE cartodb_id BETWEEN " + str(currentID) + " AND " + str(rangeEndID)

        response = requests.get(finalURL)
        rawData = response.content.decode("utf-8")

        RawDF = (spark.read
            .option("multiline", True)
            .json(sc.parallelize([rawData])))

        # now just explode out the rows struct, that's all we care about
        explodedDF = (
                    RawDF
                        .withColumn("results",explode(RawDF.rows))
                        .select ("results.*"))

        if allDataDF is not None and isinstance(allDataDF,DataFrame):
            allDataDF = allDataDF.union(explodedDF)
        else:
            allDataDF = explodedDF

        currentID = rangeEndID + 1
    
    (allDataDF
        .write
        .mode("overwrite")
        .format("delta")
        #.save("Tables/Philly-" + tbl))
        .saveAsTable("Philly_" + tbl))
    #allDataDF.createOrReplaceTempView(tbl)
    print ("completed:" + tbl)


```
